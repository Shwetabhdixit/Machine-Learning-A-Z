# Standardize Variable to identify Key features

```python
from sklearn.linear_model import LinearRegression
```

```python
model_new = LinearRegression()
```

```python
model_new.fit(X,y)
```

```python
 result = pd.DataFrame(list(zip(model_new.coef_,df.columns)),columns=['coefficient','name']).set_index('name')
    #we take together the model,coefficients for each variable and match it to corresponding columns and set the index as name
```

```python
np.abs(result).sort_values(by='coefficient',ascending=False)
```

```python
from sklearn .preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
scaler = StandardScaler()
standard_coefficient_linear_reg = make_pipeline(scaler,model_new)
#make_pipeline basically lays out the steps to be followed.
#here the model is scaled and then fit into the linear regression model
```

```python
#The idea behind StandardScaler is that it will transform your data such that 
#its distribution will have a mean value 0 and standard deviation of 1.
#Given the distribution of the data, each value in the
#dataset will have the sample mean value subtracted,
#and then divided by the standard deviation of the whole dataset.
```

```python
standard_coefficient_linear_reg.fit(X,y)
```

```python
result = pd.DataFrame(list(zip(standard_coefficient_linear_reg.steps[1][1].coef_,df.columns)),columns=['coefficient','name']).set_index('name')
np.abs(result).sort_values(by='coefficient',ascending=False)
```

```python
#NOX is no longer at the top.
#LSTAT is most significant and AGE is least significant
```

# User R2(square) to identify Key Features

```python
from sklearn.metrics import r2_score
```

```python
linear_reg = smf.ols(formula='y~CRIM+ZN+INDUS+CHAS+NOX+RM+AGE+RAD+DIS+TAX+PTRATIO+B+LSTAT',data=df)
```

```python
benchmark = linear_reg.fit()
r2_score(y,benchmark.predict(df))
```

```python
#without LSTAT

linear_reg = smf.ols(formula = 'y~CRIM+ZN+INDUS+CHAS+NOX+RM+AGE+RAD+DIS+TAX+PTRATIO+B',data=df)
lr_without_LSTAT = linear_reg.fit()
```

```python
r2_score(y,lr_without_LSTAT.predict(df)) # thus LSTAT significant
```

```python
#without age
linear_reg = smf.ols(formula = 'y~CRIM+ZN+INDUS+CHAS+NOX+RM+RAD+DIS+TAX+PTRATIO+B+LSTAT',data=df)
```

```python
Linear_reg_without_Age = linear_reg.fit()
```

```python
r2_score(y,Linear_reg_without_Age.predict(df)) #Age is not very important
```

```python
#DIS
linear_reg=smf.ols(formula = 'y~CRIM+ZN+INDUS+CHAS+NOX+RM+AGE+RAD+TAX+PTRATIO+B+LSTAT',data=df)
```

```python
linear_reg_without_DIS = linear_reg.fit()
```

```python
r2_score(y,linear_reg_without_DIS.predict(df))
```

```python
#RM
linear_reg = smf.ols(formula = 'y~CRIM+ZN+INDUS+CHAS+NOX+AGE+RAD+DIS+TAX+PTRATIO+B+LSTAT',data=df)
```

```python
linear_reg_without_RM = linear_reg.fit()
```

```python
r2_score(y,linear_reg_without_RM.predict(df))
```

```python
#clearly RM is more significant than DIS
```
