# robust regression

```python
#outliers destroy the fit of a linear regression. Slope is impacted heavily by them
```

```python
#Random Sample Consensus ( RANSAC ) Algorithm
```

```python
#RANSAC fits a model from random subsets of inliers from the complete dataset
```

```python
 #steps of the algorithms-
    #select random samples from the original data and check whether
    #the set of data is valid
    #fit a model to random subset(base_estimator.fit) and  check whether the estimated model is valid
    #classify data as outliers or inliers by calculating the residuals to the estimated model(base_estimator.predict(X)-y) - all data samples with absolute rediduals samller than the residual_threshold are considered as inliers
    #save the fitted model as the best model if number of inliers samples is maximum.
    #in case of same inliers the one with the better score is kept
```

```python
x_ran = df['RM'].values.reshape(-1,1)
y_ran = df['MEDV'].values
```

```python
from sklearn.linear_model import RANSACRegressor
ransac = RANSACRegressor()
```

```python
ransac.fit(x_ran,y_ran)
```

```python
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
```

```python
line_X = np.arange(3,10,1)
line_Y_ransac = ransac.predict(line_X.reshape(-1,1))
```

```python
sns.set(style='darkgrid',context='notebook')
plt.figure(figsize=(12,10))
plt.scatter(x_ran[inlier_mask],y_ran[inlier_mask],c='blue',marker='o',label='inliers')
plt.scatter(x_ran[outlier_mask],y_ran[outlier_mask],c='brown',marker='x',label='outliers')
plt.plot(line_X,line_Y_ransac,color='red')
plt.xlabel('avg. number of rooms per dwelling')
plt.ylabel('median value')
plt.legend(loc='upper left')
plt.show()
```

```python
ransac.estimator_.coef_
```

```python
ransac.estimator_.intercept_
```

# What does a near perfect model look like

```python
generate_random = np.random.RandomState(0)
x=10*generate_random.rand(1000)
y=3*x+ np.random.randn(1000)
plt.figure(figsize=(10,12))
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)


from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train.reshape(-1,1),y_train)

y_train_pred=model.predict(X_train.reshape(-1,1))
y_test_pred = model.predict(X_test.reshape(-1,1))
```

```python
plt.figure(figsize=(10,12))
plt.scatter(y_train,y_train-y_train_pred,c='blue',marker='o',label='Training Data')
plt.scatter(y_test,y_test-y_test_pred,c='red',marker='*',label='Testing Data',)
plt.hlines(y=0,color='black',xmin=-10,xmax=40)
plt.ylim(-15,25)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_train,y_train_pred)
```

```python
mean_squared_error(y_test,y_test_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_train,y_train_pred)
```

```python
r2_score(y_test,y_test_pred)
```

```python
#thus this is almost perfect
```
